```python
def clean_name(filename, field):
    file = pd.read_csv(filename)
    file[field] = file[field].str.replace(r'\s\d+', '', regex=True)
    file[field] = file[field].str.replace("St.", "St")
    file.to_csv(f"cleaned_{file}", index = False)
```
