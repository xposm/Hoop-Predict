```python
def create_team_performance_df(raw_df, max_daynum=132):
    # Separate winners and losers
    winners = raw_df[['Season', 'DayNum', 'WTeamID', 'WScore','WLoc'] + 
                    [c for c in raw_df.columns if c.startswith('W') and c not in ['WTeamID', 'WScore', 'WLoc']]].copy()
    losers = raw_df[['Season', 'DayNum', 'LTeamID', 'LScore','WLoc'] + 
                   [c for c in raw_df.columns if c.startswith('L') and c not in ['LTeamID', 'LScore', 'WLoc']]].copy()

    # Standardize column names
    stat_columns = ['FGM', 'FGA', 'FGM3', 'FGA3', 'FTM', 'FTA',
                    'OR', 'DR', 'Ast', 'TO', 'Stl', 'Blk', 'PF']
    winners.columns = ['Season', 'DayNum', 'TeamID', 'Score','Loc'] + stat_columns
    #You can change column names this way only when you know the order of the columns in the DF
    winners['Win'] = True 

    
    L_location_map = {'H': 'A', 'A': 'H', 'N': 'N'}
    losers['WLoc'] = losers['WLoc'].map(L_location_map)
    losers.columns = ['Season', 'DayNum', 'TeamID', 'Score', 'Loc'] + stat_columns
    losers['Win'] = False  # Add loss column


    # Combine and filter
    combined = pd.concat([winners, losers])
    return combined[combined['DayNum'] <= max_daynum].sort_values(['Season', 'DayNum', 'TeamID'])





Mregular_performance  = create_team_performance_df(MRegular)
```

```python
Mregular_performance = Mregular_performance.sort_values('TeamID')
```
